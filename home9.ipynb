```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car.csv")
df.head()
```

# Avtomashinalarni tahlil qilamiz

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",y="km_driven",data=df,kind="point",hue="fuel")
plt.xticks(rotation=75)
plt.show()
```

## yuqorida moshina egalari bosib o'tgan km va moshina energiya turi

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",y="km_driven",data=df,kind="point",hue="fuel",join=False)
plt.xticks(rotation=75)
plt.show()
```

```python
sns.set_style("white")
sns.catplot(x="owner",y="km_driven",data=df,kind="point",hue="fuel")
plt.xticks(rotation=75)
plt.show()
```

## orqa fonga ishlov berish

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",data=df,kind="count")
plt.xticks(rotation=75)
plt.show()
```

## yuqorida owner counti

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",data=df,kind="count",palette="Set1")
plt.xticks(rotation=75)
plt.show()
```

## rang berish

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",data=df,kind="count",palette="Set1",hue="transmission")
plt.xticks(rotation=75)
plt.show()
```

## carning ownerlari va transmissiyasi haqida ma'lumot

```python
sns.set_style("whitegrid")
sns.catplot(x="owner",y="km_driven",data=df,kind="point",hue="fuel")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.catplot(x="owner",y="km_driven",data=df,kind="box",hue="fuel")
plt.xticks(rotation=75)
plt.show()
```

## yuqorida car yilini matplotlibda chiqarish

## yuqorida esa car yilini pandas orqali chiqarish

```python
sns.histplot(df["year"])
plt.show()
```

```python
sns.displot(df["year"],kind="kde")
plt.show()
```

```python
sns.displot(df["year"],kde=True,bins=10)
plt.show()
```

```python
sns.displot(df["year"],kind="kde",rug=True,fill=True)
plt.show()
```

```python
sns.displot(df["year"],kind="ecdf")
plt.show()
```

## Yuqoridagilar displot orqali turli xil tushunarli infolar orqali car yilini ko'rsatish

```python
sns.regplot(data=df,x="year",y="selling_price")
plt.show()
```

## regplotda car yili

```python
sns.lmplot(data=df,x="year",y="selling_price")
plt.show()
```

```python
sns.lmplot(data=df,x="year",y="selling_price",hue="transmission")
plt.show()
```

```python
sns.lmplot(data=df,x="year",y="selling_price",col="transmission")
plt.show()
```
