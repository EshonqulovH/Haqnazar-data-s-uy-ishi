```python
df1 = pd.read_csv("insurance.csv")
df1.head()
```

```python
df1.describe()
```

# Normal distributionda outlierlar bilan ishlash

```python
sns.boxplot(df1["charges"])
```

```python
mean = df1["charges"].mean()
std = df1["charges"].std()
```

```python
x=0.5
u_limit = mean + x * std
l_limit = mean - x * std


df_new = df1[(df1["charges"] >= l_limit) & (df1["charges"] <= u_limit)]
df_new["charges"].hist(bins=20)
```

```python
sns.boxplot(df_new["charges"])
```

# Skewed distiributionda outlearlarga ishlov berish

```python
sns.boxplot(df1["bmi"])
```

```python
Q1 = df1['bmi'].quantile(0.25)
Q3 = df1['bmi'].quantile(0.75)
IQR = Q3 - Q1
```

```python
df_filtered = df1[(df1['bmi'] >= (Q1 - 1.5 * IQR)) & (df1['bmi'] <= (Q3 + 1.5 * IQR))]
df_filtered["bmi"].hist(bins=15)
```

```python
sns.boxplot(df_filtered["bmi"])
```
