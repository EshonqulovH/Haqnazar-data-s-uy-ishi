# 25000 ming qatorda Nan bo'lganlarni tashavorish va yangi faylga olish

```python
x = 25000
df = df.loc[:, df.isna().sum() <= x]
```

# Bu yerda Nan qiymatlarga ustun boyicha sonlarni medianasini berib chiqdik

```python
df["yacEPA"] = df["yacEPA"].fillna(df["yacEPA"].median())
df
```

```python
df["airEPA"] = df["airEPA"].fillna(df["airEPA"].median())
df
```

# Bu yerda objekt bo'lgani uchun modani Nan qiymatlarga berib chiqdik

```python
y = df["Passer"].mode()[0]
df["Passer"] = df["Passer"].fillna(value=y)
df
```

```python
x = df["PassOutcome"].mode()[0]
df["PassOutcome"] = df["PassOutcome"].fillna(value=x)
df
```
