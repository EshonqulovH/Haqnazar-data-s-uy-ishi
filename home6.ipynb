```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world_pop.csv")
df.head(10)
```

# Aholisi eng ko'p 10 ta mamlakatni ajratib tahlil qilamiz

```python
fig, ax = plt.subplots()
ax.set_title("10 most populous countries(2023)")
ax.plot(df["cca3"].head(10),df["pop2023"].iloc[:10],label="2023 year",marker="v",color="green")
ax.set_xlabel("Country")
ax.set_ylabel("Populations(billon)")
ax.legend()
plt.show()
```

## Yuqorida 10 ta davlatning 2023 yil aholisi  haqida ma'lumot

```python
fig, ax = plt.subplots()
ax.set_title("10 most populous countries(1980-2050)")
ax.plot(df["cca3"].head(10),df["pop1980"].iloc[:10],label="1980 year",marker="v",color="blue",linestyle="--")
ax.plot(df["cca3"].head(10),df["pop2050"].iloc[:10],label="2050 year",marker="v",color="green",linestyle="--")
ax.set_xlabel("Country")
ax.set_ylabel("Populations(billon)")
ax.legend()
plt.show()
```

## Yuqorida 10 ta davlatning 1980 va 2050 yillardagi aholisi o'sishi statistikasi

```python
plp = pd.read_csv("world_pop.csv",index_col=0)
plp = plp.iloc[:10]
fig, ax = plt.subplots()
ax.bar(plp.index,plp["pop2010"],label="2010")
ax.bar(plp.index,plp["pop2023"],bottom=plp["pop2010"],label="2023")
ax.bar(plp.index,plp["pop2030"],bottom=plp["pop2010"]+plp["pop2023"],label="2030")
ax.bar(plp.index,plp["pop2050"],bottom=plp["pop2010"]+plp["pop2023"]+plp["pop2030"],label="2050")
ax.set_xticklabels(plp.index,rotation=75)
ax.set_xlabel("Country")
ax.set_ylabel("Population(billion)")
ax.legend()
plt.show()
```

# Train csv fayli bilan Erkaklar va Ayollarning yoshlarini tahlil qilish

```python
df1 = pd.read_csv("train.csv",index_col=0)
df1.head()
```

```python
#erkak va ayollarni filtrlab o'zgaruvchiga ajratib olish
mens = df1[df1["Sex"]=="male"]
womens = df1[df1["Sex"]=="female"]
```

```python
fig, ax = plt.subplots()
ax.bar("Erkaklar",mens["Age"].mean())
ax.bar("Ayollar",womens["Age"].mean())
ax.set_ylabel("AVG age")
plt.show()
```

## yuqoridagi ma'lumot erkak va ayollarning o'rtacha yoshi

```python
fig, ax = plt.subplots()
ax.bar("Erkaklar",mens["Age"].mean(),yerr=mens["Age"].std())
ax.bar("Ayollar",womens["Age"].mean(),yerr=womens["Age"].std())
ax.set_ylabel("AVG age")
plt.show()
```

## yuqoridagi ma'lumot erkak va ayollarning avg yoshini standart og'ishi

```python
fig, ax = plt.subplots()
ax.hist(mens["Age"])
ax.hist(womens["Age"])
ax.set_xlabel("Year old")
ax.set_ylabel("# of observations")
plt.show()
```

## yuqoridagi malumot erkak va ayollarning yoshlari "histogram" grafikasi asosida

```python
fig, ax = plt.subplots()
ax.hist(mens["Age"],label="Erkaklar",bins=5)
ax.hist(womens["Age"],label="Ayollar",bins=5)
ax.set_xlabel("Year old")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## yuqorida Gistogrammalarni sozlash: qutilar sonini o'rnatish

```python
fig, ax = plt.subplots()
ax.hist(mens["Age"],label="Erkaklar",bins=5,histtype="step")
ax.hist(womens["Age"],label="Ayollar",bins=5,histtype="step")
ax.set_xlabel("Year old")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## Yuqorida qadamlar gistogrammasi erkaklar va ayollar yoshining

```python
fig, ax = plt.subplots()
ax.boxplot([mens["Fare"],womens["Fare"]])
ax.set_xticklabels(["Erkaklar","Ayollar"])
ax.set_ylabel("Funt")
plt.show()
```

```python
seatter_weather=pd.read_csv("seattle_weather.csv")
fig, ax = plt.subplots(2, 1)
station1 = seatter_weather[seatter_weather["STATION"] == "USC00456295"]
station2 = seatter_weather[seatter_weather["STATION"] == "USC00454169"]
ax[0].scatter(seatter_weather["MLY-PRCP-NORMAL"], seatter_weather["MLY-TAVG-NORMAL"])
std1 = station1["MLY-TAVG-NORMAL"].std()
std2 = station2["MLY-TAVG-NORMAL"].std()
ax[1].bar("USC00456295", station1["MLY-TAVG-NORMAL"].mean(), yerr=std1)
ax[1].bar("USC00454169", station2["MLY-TAVG-NORMAL"].mean(), yerr=std2)
ax[0].set_xlabel("PRCP")
ax[0].set_ylabel("TAVG")
ax[1].set_xlabel("station")
ax[1].set_ylabel("TAVG")
plt.show()
```
