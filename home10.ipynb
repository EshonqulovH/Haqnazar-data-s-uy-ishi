```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("car.csv")
df.head()
```

```python
sns.stripplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.swarmplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.boxplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.violinplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.boxenplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.barplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.pointplot(data=df,x="owner",y="selling_price",hue="owner")
plt.xticks(rotation=75)
plt.show()
```

```python
sns.regplot(data=df,x="year",y="selling_price",marker="+")
plt.show()
```

```python
sns.residplot(data=df,x="year",y="selling_price")
plt.show()
```

```python
sns.regplot(data=df,x="year",y="selling_price",order=2)
plt.show()
```

```python
sns.residplot(data=df,x="year",y="selling_price",order=2)
plt.show()
```

```python
sns.regplot(data=df,x="year",y="selling_price",x_jitter=.1,order=2)
plt.show()
```

```python
sns.regplot(data=df,x="year",y="selling_price",x_estimator=np.mean,order=2)
plt.show()
```

```python
sns.regplot(data=df,x="year",y="selling_price",order=2,x_bins=5)
plt.show()
```

```python
df_crosstab=pd.crosstab(df["year"],df["owner"],values=df["selling_price"],aggfunc="mean")
```

```python
sns.heatmap(pd.crosstab(df["year"],df["owner"],values=df["selling_price"],aggfunc="mean"))
```

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=False, linewidths=.5)
```

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=True, linewidths=1)
```
