```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cinema.csv")
df.head()
```

# Budget ustuni o'zgarishdan oldin

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```

```python
df["Budget"].describe()
```

# Standartisation formulasi bilan budget ustuni sonlarini o'zgartirish

```python
df["Budget"] = (df["Budget"]-df["Budget"].mean())/df["Budget"].std()
```

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```

# Min-Max scalling formulasi bilan budget ustuni sonlarini o'zgartirsh

```python
df["Budget"]=(df["Budget"]-df["Budget"].min())/(df["Budget"].max()-df["Budget"].min())
df.head()
```

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```

# Mean normalisation formulasi bilan ustun sonlarini o'zgartirish

```python
df["Budget"]=(df["Budget"]-df["Budget"].mean())/(df["Budget"].max()-df["Budget"].min())
df.head()
```

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```

# Robust scalling bilan ustundagi sonlarni o'zgartirish

```python
df["Budget"]=(df["Budget"]-df["Budget"].median())/(df["Budget"].quantile(0.75)-df["Budget"].quantile(0.25))
df.head()
```

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```

# Maximum absolut scalling bn ustun sonlarini o'zgartirish

```python
df["Budget"]=df["Budget"]/df["Budget"].abs().max()
df.head()
```

```python
sns.histplot(df["Budget"],kde=True,bins=10)
plt.show()
```
